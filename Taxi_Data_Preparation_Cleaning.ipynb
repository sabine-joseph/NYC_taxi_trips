```python
import pandas as pd
import numpy as np
```

```python
NYC_taxi_data = pd.read_csv('Desktop/Taxi_Task/yellow_tripdata_2015-06.csv')
size_before_cleaning = len(NYC_taxi_data)
NYC_taxi_data.head()
```

```python
# any nulls in df?
NYC_taxi_data.isnull().values.any()
```

```python
# descriptive stats 
NYC_taxi_data.describe()
```

```python
# 'trip_distance' 
NYC_taxi_data = NYC_taxi_data[((NYC_taxi_data['trip_distance'] > 0) & (NYC_taxi_data['trip_distance'] < 10))]

size_after_cleaning = len(NYC_taxi_data)

percentage_data_remaining = float(size_after_cleaning)/float(size_before_cleaning)*100
print percentage_data_remaining
```

```python
#'tip_amount'
count_neg_values = (NYC_taxi_data['tip_amount'] < 0).sum()

NYC_taxi_data = NYC_taxi_data.loc[NYC_taxi_data['tip_amount'] >= 0] # remove negative values

# outliers:
outlier_cut_off = round(NYC_taxi_data['tip_amount'].mean()+3*NYC_taxi_data['tip_amount'].std())

# count outliers
num_tip_outliers = (NYC_taxi_data['tip_amount'] < outlier_cut_off).sum()
percentage_data_with_outliers_removed = float(size_after_cleaning-num_tip_outliers)/float(size_before_cleaning)*100

# -> didn't remove outliers
```

```python
# coordinates
NYC_taxi_data = NYC_taxi_data[((NYC_taxi_data['dropoff_longitude'] != 0) & (NYC_taxi_data['dropoff_latitude'] != 0))]
NYC_taxi_data = NYC_taxi_data[((NYC_taxi_data['pickup_longitude'] != 0) & (NYC_taxi_data['pickup_latitude'] != 0))]

print min(NYC_taxi_data['dropoff_longitude'])
print min(NYC_taxi_data['pickup_longitude'])
print max(NYC_taxi_data['dropoff_latitude'])
print max(NYC_taxi_data['pickup_latitude'])
```

```python
NYC_taxi_data = NYC_taxi_data[(NYC_taxi_data['dropoff_longitude'] >= -86)]
percentage_data_remaining = len(NYC_taxi_data)/float(size_before_cleaning)*100
print percentage_data_remaining
```

```python
# write cleaned data to csv
NYC_taxi_data.to_csv('Desktop/Taxi_Task/cleaned_NYC_taxi_data_with_geo.csv')
```
